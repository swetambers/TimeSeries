##### Models for time series analysis:

Chosen model should be able to account for one or more of the internal structures that might be present (i.e 4 components - General trend, Seasonality, Cyclical movements, Unexpected variations)

- <b>Zero mean models</b> : 
    - Constant mean and constant variance and shows no predictable trends or seasonality.
    - Independent and identically distributed (iid).
    - Random noise around a fixed mean: Modeled by a normal distribution of mean zero and variance σ^2, which is assumed to be the irreducible error of the model and hence treated as a random noise. 

 <br>
 <br>

- <b>Random walk</b> : 
    - Given as a sum of n iids aka Zero mean models.
    - Reduced to zero mean model by taking differences of the observations from two consecutive time indices. 

 <br>
 <br>
    
- <b>Trend models</b>:
    - Captures the long run trend in the time series that can be fitted as linear regression of the time index.
    - If time series does not exhibit any periodic or seasonal fluctuations, it can be expressed just as the sum of the trend and the zero mean model as xt = μ(t) + yt, where μ(t) is the time-dependent long run trend of the series.
    - Can be liner or non linear in 't'
    - Residuals xt - μ(t) of the trend model is considered to the irreducible noise and as realization of the zero mean component yt

 <br>
 <br>
    
- <b>Seasonality models</b>:
    - Periodic and repetitive fluctuations in a time series and hence are modeled as sum of weighted sum of sine waves of known periodicity.
    - Harmonic regression model as they attempt to fit the sum of multiple sin waves. 

 <br>
 <br>
    

A zero sum model represents irreducible error of the system and all of other three models aim to transform a given time series to the zero sum models through suitable mathematical transformations.

Time series analysis in the following four steps:
- Visualize the data at different granularities of the time index to reveal long run trends and seasonal fluctuations
- Fit trend line capture long run trends and plot the residuals to check for seasonality or irreducible error
- Fit a harmonic regression model to capture seasonality
- Plot the residuals left by the seasonality model to check for irreducible error


```python
import os
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
zero_mean_series = np.random.normal(loc=0.0, scale=1., size=100)
```

```python
plt.figure(figsize=(5.5, 5.5))
g = sns.tsplot(zero_mean_series)
g.set_title('Zero mean model')
g.set_xlabel('Time index')
```

```python
random_walk = np.cumsum(zero_mean_series)
```

```python
plt.figure(figsize=(5.5, 5.5))
g = sns.tsplot(random_walk)
g.set_title('Random Walk')
g.set_xlabel('Time index')
```
